### Import Dataset from csv or txt and check variance in the columns

```python
import os
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold
from matplotlib import pyplot

# define the location of the dataset
#os.chdir("D:\Investing Papers\Machine Learning")
#cwd = os.getcwd()


# load the dataset as csv or txt file, selecting the decimal symbol, headers, and delimiter
#df = pd.read_csv(r'valeursfoncieres-2018.txt', , decimal=",", delimiter="|", header=None)

from pandas import read_csv
# define the location of the dataset
path = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/oil-spill.csv'
# load the dataset
df = read_csv(path, header=None)

print("Original Shape",df.shape)
# summarize the number of unique values in each column
print(df.nunique())

# record columns to delete - with only one unique value
counts = df.nunique()
to_del = [i for i,v in enumerate(counts) if v == 1]
print("Columns to delete",to_del)
# drop useless columns
df.drop(to_del, axis=1, inplace=True)
print("Unique columns",df.shape)

#record columns to delete - with less than 1% unique values
counts = df.nunique()
to_del = [i for i,v in enumerate(counts) if (float(v)/df.shape[0]*100) < 1]
print(to_del)
#drop useless columns
df.drop(to_del, axis=1, inplace=True)
print("Varying Columns",df.shape)

# summarize again the number of unique values in each column
print(df.nunique())

#The below approach can be used instead of the above to narrow down the dataset for low-variance variables
data = df.values
X = data[:, :-1]
y = data[:, -1]
thresholds = np.arange(0.0, 0.55, 0.05)

results = list()
for t in thresholds:
    # define the transform
    transform = VarianceThreshold(threshold=t)
    # transform the input data
    X_sel = transform.fit_transform(X)
    # determine the number of input features
    n_features = X_sel.shape[1]
    print('>Threshold=%.2f, Features=%d' % (t, n_features))
    # store the result
    results.append(n_features)
# plot the threshold vs the number of selected features
pyplot.plot(thresholds, results)
pyplot.show()

#This acheives the same as above without converting the dataframe to a numpy array. Note that you need to exclude the
#target variable from this selector

#Extract target variable from dataframe
y = df[49]
df.drop(49, axis=1, inplace=True)

#Run variance threshold selector
def variance_threshold_selector(data, threshold=0.5):
    selector = VarianceThreshold(threshold)
    selector.fit(data)
    return data[data.columns[selector.get_support(indices=True)]]

df = variance_threshold_selector(df, 0.0)

#add back the value of y as its original column name in the dataframe and show the dataframe
df[49] = y
df
```

```python
#This code is to use the knn-imputation approach to missing data

import pandas as pd
import numpy as np
import os
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

# define the location of where to load the data, if from a drive
os.chdir("D:\Investing Papers\Machine Learning")
cwd = os.getcwd()

path = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/horse-colic.csv'
# load the dataset - may need to change header value to 0 if there are headers in the first row of the csv file
df = pd.read_csv(path, header=None, na_values='?')
print(df.head())

#Convert dataframe to numpy array
data = df.values

#Create a list of columns which should be in X data (i.e. all but the one target variable)
ix = [i for i in range(data.shape[1]) if i != 23]

#Define X and y data
X, y = data[:, ix], data[:, 23]

#Scale the features which are not already binary fields
#First we fix the fact that this specifc dataset uses 1 and 2 instead of 0 and 1 for categorical data
fixcols = [0,26]
for i in fixcols:
    X[:,i][X[:,i] == 1] = 0
    X[:,i][X[:,i] == 2] = 1

#Standardize the data where the algos require data have similar scale and variance
scaler = StandardScaler()
X = scaler.fit_transform(X)

#Alternatively, scale data between 0 and 1 where only scaling is needed but variance not important
#scaler = MinMaxScaler(feature_range=(0,1))
#X = scaler.fit_transform(X)

# print total missing fields
print('Missing: %d' % sum(np.isnan(X).flatten()))

# define imputer
imputer = KNNImputer()

# fit on the dataset
imputer.fit(X)

# transform the dataset
Xtrans = imputer.transform(X)

# print total missing
print('Missing: %d' % sum(np.isnan(Xtrans).flatten()))

#Combine transformed X data and y data into a dataframe
df1 = pd.DataFrame(np.column_stack((Xtrans,y)))

#Export the original dataframe and amended dataframe for comparison
os.chdir("D:\Investing Papers\Machine Learning\DataOutputs")
cwd = os.getcwd()
df.to_csv('DF.txt',sep=';')
df1.to_csv('DF1.txt',sep=';')
```

### One-hot encoding

```python
import pandas as pd
import numpy as np
import os

os.chdir("D:\Investing Papers\Machine Learning")
cwd = os.getcwd()

#open the data file
df = pd.read_csv(r'valeursfoncieres-2018.txt', decimal=",", delimiter="|", usecols = [9,10,16,18,19,20,21,22,35,38,39,41,42], 
                 dtype = {"Nature mutation" : "category","Valeur fonciere" : "float64","Code postal" : "category",
                          "Code departement":"object","Section" : "category", "Code type local" : "category",
                          "Surface reelle bati" : "float64", "Nombre pieces principales" : "float64",
                          "Nature culture speciale" : "category", "Surface terrain" : "float64"},)

#Add index
df['Index'] = df['Code departement'].astype(str) + '-' + df['Code commune'].astype(str) + '-'\
+ df['Prefixe de section'].astype(str) + '-' + df['Section'].astype(str) + '-' + df['No plan'].map(str)

df.set_index('Index')

#Drop columns only imported for the index creation
df.drop(['Code departement', 'Code commune', 'Prefixe de section','No plan'], axis=1, inplace=True)

#Rename column headers
df.columns = ["NatureMutation","ValeurFonciere","CodePostal","Section","CodeTypeLocal",
              "SurfaceReelleBati", "NombrePiecesPrincipales","NatureCultureSpeciale","SurfaceTerrain","Index"]

#Replace NA values with special category and 0, as appropriate
df['NatureCultureSpeciale'] = df['NatureCultureSpeciale'].cat.add_categories("Neant")
df['NatureCultureSpeciale'] = df['NatureCultureSpeciale'].fillna("Neant")
df['SurfaceTerrain'] = df['SurfaceTerrain'].fillna(0)

#remove rows with data gaps
df.dropna(axis=0,how='any',thresh=None, subset=None,inplace=True)

#Remove all data except Paris, Lille, Strasbourg
postcodearray = ['75000','75001','75002','75003','75004','75005','75006','75007','75008','75009','75010',
                 '75011','75012','75013','75014','75015','75016','75017','75018','75019','75020','69000',
                 '69001','69002','69003','69004','69005','69006','69007','69008','69009','59000','59160',
                 '59260','59777','59800','67000','67100','67200']
df = df.loc[df['CodePostal'].isin(postcodearray)]

#Remove all that are not houses or apartments
filtrebien = ['1','2']

df = df.loc[df['CodeTypeLocal'].isin(filtrebien)]

#Add city name
df['Ville'] = np.where(df['CodePostal'].str.startswith('75'),'Paris',
                            np.where(df['CodePostal'].str.startswith('69'),'Strasbourg',
                                    'Lille'))
df['Ville'] = df['Ville'].astype('category')

#Remove unused categories now that we have narrowed down the cities and type of building
df['CodePostal'] = df.CodePostal.cat.remove_unused_categories()
df['NatureCultureSpeciale'] = df.NatureCultureSpeciale.cat.remove_unused_categories()
df['Section'] = df.Section.cat.remove_unused_categories()
df['CodeTypeLocal'] = df.CodeTypeLocal.cat.remove_unused_categories()

#perform one-hot encoding on the necessary columns
#note that if columns not specified, all columns of dtype 'object' or 'category' will be encoded, so if this is done
#during the import as above, it should not be necessary to specify the columns
features = pd.get_dummies(df, columns = ['NatureMutation','CodePostal', 'Section', 'CodeTypeLocal', 
                                                 'NatureCultureSpeciale', 'Ville'],
                                  prefix=['NatureMutation','CodePostal', 'Section', 'CodeTypeLocal',
                                          'NatureCultureSpeciale','Ville'])
print(features.head())

#Create the X and y data for the model
X = features.drop(['ValeurFonciere'],axis=1)
y = features['ValeurFonciere']

#View the X and y data in csv if necessary
#X.to_csv(r'D:\Investing Papers\Machine Learning\DataOutputs\X.csv', index = None, header=True, sep=';')
#y.to_csv(r'D:\Investing Papers\Machine Learning\DataOutputs\y.csv', index = None, header=True, sep=';')
```

### Handling Outliers

```python
#Use an isolation forest to remove outliers from the dataset

from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
import numpy as np

# load the dataset and convert to a numpy array
path = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv'
df = read_csv(path)
dataset = df.to_numpy()

#Split the data into features and y values, then split into train and test sets
X, Y = dataset[:,0:8], dataset[:,8]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)

#Show the original dataset dimensions
print('\033[1m','X and Y Shapes - Original\n','\033[0m',np.shape(X_train),np.shape(y_train),'\n')

#Run the Isolation Forest to remove outliers
iso = IsolationForest(contamination='auto')
ypred = iso.fit_predict(X_train) #-1 is the value given if it is an outlier or 1 if not an outlier
mask = ypred != -1

#Remove the outliers from the training data
X_train, y_train = X_train[mask,:], y_train[mask]

#Show the amended dataset dimensions and the number of outliers removed
print('\033[1m','X and Y Shapes - Amended\n','\033[0m',np.shape(X_train),np.shape(y_train),'\n')
print('\033[1m','Number of Outliers Removed\n','\033[0m',np.count_nonzero(ypred == -1),'\n')

```

```python
#Use the minimum covariance determinant to remove outliers from the dataset
#Note - we should only use this if we are sure the features of the 
#dataset have a Gaussian distribution. 
#The Elliptic Envelope function provides the contamination argument
#that defines the expected ratio of outliers to be observed in practice.
#This will need to be defined on a case-by-case basis with trial and error.

from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.covariance import EllipticEnvelope
import numpy as np

# load the dataset and convert to a numpy array
path = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv'
df = read_csv(path)
dataset = df.to_numpy()

#Split the data into features and y values, then split into train and test sets
X, Y = dataset[:,0:8], dataset[:,8]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)

#Show the original dataset dimensions
print('\033[1m','X and Y Shapes - Original\n','\033[0m',np.shape(X_train),np.shape(y_train),'\n')

#Run the Isolation Forest to remove outliers
ee = EllipticEnvelope()
ypred = ee.fit_predict(X_train)
mask = ypred != -1

#Remove the outliers from the training data
X_train, y_train = X_train[mask,:], y_train[mask]

#Show the amended dataset dimensions and the number of outliers removed
print('\033[1m','X and Y Shapes - Amended\n','\033[0m',np.shape(X_train),np.shape(y_train),'\n')
print('\033[1m','Number of Outliers Removed\n','\033[0m',np.count_nonzero(ypred == -1),'\n')
```
