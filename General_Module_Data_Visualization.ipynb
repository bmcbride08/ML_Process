```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv'
# load the dataset - may need to change header value to 0 if there are headers in the first row of the csv file
df = pd.read_csv(path, header=None)
df.columns = ["Number_of_times_pregnant","Plasma_glucose_concentration","Diastolic_blood_pressure",
              "Triceps_skin_fold_thickness","2-Hour_serum_insulin",
              "BMI", "Diabetes_pedigree_function","Age","ClassVariable"]

pd.plotting.scatter_matrix(df, alpha=0.8, figsize=(25,25))

#Same visualization but with seaborn
plt.figure(figsize=(25,25))
plot = sns.pairplot(df,hue = 'ClassVariable',diag_kind = "hist",kind = "scatter",palette = "husl")


```
